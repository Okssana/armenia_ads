```python

import pandas as pd
from bs4 import BeautifulSoup
import cloudscraper # https://pypi.org/project/cloudscraper/
```

```python
######### 1. Get the page #1 data #########

scraper = cloudscraper.create_scraper()
page = scraper.get("https://www.list.am/ru/category/199").text
soup = BeautifulSoup(page, 'html.parser')

# Find the parent div
parent_div = soup.find('div', {'id': 'contentr'})

# Find all child divs with class 'gl' under parent div
child_divs = parent_div.find_all('div', class_='dl')#.find('div', class_='gl')

data = []

for div in child_divs:
    a_tags = div.find_all('a')
    print(a_tags)
    #link = a['href']
    for a in a_tags:
        link = a['href']
        print(link)
        # only links with 'item' in the url
        if 'item' in link:
            my_url = 'https://www.list.am'+link
            #print(my_url)

            description_1 = a.find('div', class_='l')
            description_2 = a.find('div', class_='at')
            price = a.find('div', class_='p')

            data.append({
                'Description_1': description_1.text if description_1 else None,
                'Description_2': description_2.text if description_2 else None,
                'Price': price.text if price else None,
                'Link': my_url if my_url else None
            })

df = pd.DataFrame(data)

#print(df.to_csv('data.csv', index=False))
display(df)

```

```python
# df['Link'] contains the inner URL similar to https://www.list.am/ru/item/19684718

# Extract the item id from the URL
df['url_item_id'] = df['Link'].str.replace('https://www.list.am/ru/item/', '')
df
```

```python

# Run through the list of URLs and scrape the data

data_list = []

for url_item in df['Link'][0:3]:
    print(url_item)
    page = scraper.get(url_item).text
    soup = BeautifulSoup(page, 'html.parser')
    #print(soup.prettify())


    # Create a dictionary to store the scraped data
    data = {}

    # Find the parent div
    parent_div_rent = soup.find('div', class_='vi')

    # Find all child divs with class 'c' under the parent div
    child_divs_rent = parent_div_rent.find_all('div', class_='c')

    text_description = soup.find('div', class_='body')
    #print(text_description.text)

    # Iterate over each 'c' div
    for div in child_divs_rent:
        # Extract the 't' and 'i' divs
        title_div = div.find('div', class_='t')
        info_div = div.find('div', class_='i')

        # If both divs are found, add them to the data dictionary
        if title_div and info_div:
            key = title_div.text.strip()  # .strip() is used to remove leading/trailing whitespaces
            value = info_div.text.strip()
            
            # Convert the title to a variable-friendly format (all lowercase, spaces replaced with underscores)
            key = key.lower().replace(' ', '_')

            # Store the data
            data[key] = value
    
    
    data['text_description'] = text_description.text
    # Add the url_item to the data dictionary
    data['url_item'] = url_item
    # Add the data dictionary to the data list
    data_list.append(data)
```

```python
# Convert data_list to a pandas DataFrame
dfd = pd.DataFrame(data_list)

# Display the DataFrame
display(dfd)

# In column url_item, replace the url with the item id
dfd['url_item_id'] = dfd['url_item'].str.replace('https://www.list.am/ru/item/', '')
display(dfd)
```

```python
# Join dfd['url_item_id'] with df['url_item_id']
df_final = df.merge(dfd, on='url_item_id', how='outer')
display(df_final)

df_final.to_csv('final_data.csv', index=False)
```
